```python
import random
import math
```

```python
#Устанавливаем изначальную температуру
def Temperature(StartTempr,i):
    return(StartTempr*(0.1/(i+1)))
```

```python
#Устанавливаем вероятность перехода к состоянию с наибольшей величиной
def stochast(dE,Ti):
    return math.exp(-dE/Ti)
```

```python
#ВЫдача решений по вероятности
def solution(prb):
    realSol=1.0-prb
    if random.uniform(0.0, 1.0)-realSol<0:
        return(1)
    else: return(0)    
```
